## Analyze Pandas Dataframe

```python
import pandas as pd

df = pd.DataFrame({'Name':['John', 'Alice', 'Pam', 'Hope'], 'Age':[20.5, 39.1, 20.5, 36.0], 'Religious': ['yes', 'yes', 'no', 'yes']})

df.head()
```

```python
df.info()

```

```python
# view non-numeric information
for col in df.columns:
    col_type = df[col].dtype
    #print(col, col_type)
    if col_type == 'object':
        print(col, '\n', df[col].value_counts(), f'\nTotal Values: {len(df[col].value_counts())}', '\n\n')

```

```python
# view numeric information
df.describe()
```

```python
# plot numeric information
import matplotlib.pyplot as plt

# extra code â€“ the next 5 lines define the default font sizes
plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)

df.hist(bins=50, figsize=(12, 8))
plt.show()
```
