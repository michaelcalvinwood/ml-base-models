```python
# There is no header. Therefore load it as such
import pandas as pd
df = pd.read_csv('arrhythmia.data', header=None)
data = df[[0, 1, 2, 3, 4, 5]] # load the first six columns
data.columns = ['age', 'sex', 'height', 'weight', 'QRS Duration', 'P-R Interval']
data.head()


```

```python
# get a histogram of the values in each column
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 15] # enlarge the figure size because they are small by default
data.hist(); # use a semicolon to suppress the return value
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(data);

```
