### Hours of Study and Test Scores

```python
import pandas as pd
def test_scores_from_hours_of_study():
  return pd.DataFrame({
      "Hours of Study": [20, 23, 25, 28, 30, 37, 40, 43, 46],
      "Test Score": [45, 51, 55, 61, 65, 79, 85, 91, 97]
  })

df = test_scores_from_hours_of_study()
df.head()

import random
import numpy as np
def formulaic_test_scores_from_hours_of_study(num_samples):
  hours = np.array(random.sample(range(0, 45), num_samples))
  scores = hours * 2.2
  return pd.DataFrame({
      "Hours of Study": hours,
      "Test Score": scores
  })

df = formulaic_test_scores_from_hours_of_study(10)
df.head()
```
